```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

data = pd.merge(ratings, movies, on='movieId')
```

```python
data
```

```python
user_item_matrix = data.pivot_table(index='userId', columns='title', values='rating')
```

```python
user_item_matrix.fillna(0, inplace=True)
```

```python
user_item_matrix
```

```python
user_similarity = cosine_similarity(user_item_matrix)
```

```python
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
user_similarity_df
```

```python
def get_top_n_similar_users(user_id, n=10):
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:n+1]
    return similar_users
```

```python
def get_movie_recommendations(user_id, user_item_matrix, n_recommendations=10):
    similar_users = get_top_n_similar_users(user_id)
    similar_users_ratings = user_item_matrix.loc[similar_users]

    user_similarity_scores = user_similarity_df[user_id].loc[similar_users]
    weighted_ratings = similar_users_ratings.T.dot(user_similarity_scores) / user_similarity_scores.sum()

    user_rated_movies = user_item_matrix.loc[user_id]
    recommendations = weighted_ratings[~user_rated_movies.index.isin(user_rated_movies[user_rated_movies > 0].index)]
   
    top_n_recommendations = recommendations.sort_values(ascending=False).head(n_recommendations).index
    return top_n_recommendations

user_id = 200
recommendations = get_movie_recommendations(user_id, user_item_matrix)
print(f"Top 10 recommendations for user {user_id}:")
for idx, movie in enumerate(recommendations, 1):
    print(f"{idx}. {movie}")
```
